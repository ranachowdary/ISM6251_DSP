# Initial setup

# Let's import a number of libraries that we expect to use

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(1) # set this to ensure the results are repeatable. 

sample_size = 1000
```

# Let's define a hidden linear relationship/model
First let's define our relationship (normally, this is hidden, but since we are creating the data, we will need to identify this)

```python
# we will define a linear model with the following parameter values (arguements)
b1 = 30.5 # slope for x
b0 = 20 # intercept
```

# FIRST COPY INCREASING THE NOISE 

```python
# we will generate the x data using a normal distribution
x_mean = 100
x_stdev = 10
X = np.round(np.random.normal(x_mean, x_stdev, sample_size),2)

# we will generate the y data using our 'hidden function' and the x data
y = b0 + b1 * np.sin(X) 

# we will add some noise to the y data
e_mean = 0
e_stdev = 15
e = np.round(np.random.normal(e_mean, e_stdev, sample_size), 2)
y = y + e
```

# Let's plot this sample data using a scatter plot

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(X, y, color='red')

ax.set_xlabel('input')
ax.set_ylabel('target')
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame({'X': X, 'y': y})
df.head(10)
```

```python
df.to_csv('c02_dataset.csv', index=False)
```

# Load the given data that we will model

```python
df = pd.read_csv('c02_dataset.csv')

df.head(10)
```

Seperate out the input m(X) and the target (y) 

```python
X = df[['X']]
y = df[['y']]
```

# Explore the given data 

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(X, y, color='red')

ax.set_xlabel('input')
ax.set_ylabel('target')
plt.tight_layout()
plt.show()
```

# Fit the model

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
_ = lin_reg.fit(np.sin(X),y)
```

# Analyze the linear regression model's performance

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(X, y, color='red')
ax.scatter(X,lin_reg.predict(np.sin(X)),color='blue')


ax.set_title('Linear Regression')
ax.set_xlabel('input')
ax.set_ylabel('target')
plt.tight_layout()
plt.show()
```

```python
b0 = lin_reg.intercept_[0]
b1 = lin_reg.coef_[0][0]
r2 = lin_reg.score(X, y)
print(f"Y = {b0:.2f} + {b1:.2f}sin(x)")
print(f"R^2: {lin_reg.score(np.sin(X), y):.3f}") 
```
